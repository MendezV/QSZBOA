```python
def func_prime_1(x, y_1, y_2):
    return y_2

def alpha(x):
    return sin(x)

beta3=-30.0
def beta(y_1):
    #return -3*y_1+y_1**2+y_1**3
    return -3.0*beta3*0.5+2*y_1+3*beta3*y_1**2

epsilon=0.01
def func_prime_2(x, y_1, y_2):
    return -y_1-epsilon*alpha(x)*beta(y_1)

def RungeKuttaFourthOrderStep(x_old, y1_old, y2_old):
    
    k_1_prime1 = func_prime_1(x_old,y1_old, y2_old)
    k_1_prime2 = func_prime_2(x_old,y1_old, y2_old)
    
    #first step
    x1 = x_old+ (h/2.0)
    y1_1 = y1_old + (h/2.0) * k_1_prime1
    y2_1 = y2_old + (h/2.0) * k_1_prime2
    k_2_prime1 = func_prime_1(x1, y1_1, y2_1)
    k_2_prime2 = func_prime_2(x1, y1_1, y2_1)
    
    #second step
    x2 = x_old + (h/2.0)
    y1_2 = y1_old + (h/2.0) * k_2_prime1
    y2_2 = y2_old + (h/2.0) * k_2_prime2
    k_3_prime1 = func_prime_1(x2, y1_2, y2_2)
    k_3_prime2 = func_prime_2(x2, y1_2, y2_2)
    
    
    #third
    x3 = x_old + h
    y1_3 = y1_old + h * k_3_prime1
    y2_3 = y2_old + h * k_3_prime2
    k_4_prime1 = func_prime_1(x3, y1_3, y2_3)
    k_4_prime2 = func_prime_2(x3, y1_3, y2_3)
    
    #fourth step
    average_k_1 = (1.0/6.0)*(k_1_prime1 + 2.0*k_2_prime1 + 2.0*k_3_prime1 + k_4_prime1)
    average_k_2 = (1.0/6.0)*(k_1_prime2 + 2.0*k_2_prime2 + 2.0*k_3_prime2 + k_4_prime2)
    
    x_new = x_old + h
    y_1_new = y1_old + h * average_k_1
    y_2_new= y2_old + h * average_k_2
    return x_new, y_1_new, y_2_new
```

```python
#we define the arrays holding the two variables
h=0.01
min_x = 0.0
max_x = 2*pi
n_points = int((max_x-min_x)/h)
x = zeros(n_points)
y_1 = zeros(n_points)
y_2 = zeros(n_points)

x[0]   = min_x
y_1[0] = -1/sqrt(2)
y_2[0] = 1/sqrt(2)   


for i in range(1,n_points):
    x[i],y_1[i],y_2[i] = RungeKuttaFourthOrderStep(x[i-1], y_1[i-1], y_2[i-1])
```

```python
def Evol(x0, y_1_0, y_2_0,n_point):
    xpre, y1pre, y2pre=x0, y_1_0, y_2_0
    xfut, y_1_fut, y_2_fut=0,0,0
    for i in range(1,n_points):
        xfut, y_1_fut, y_2_fut = RungeKuttaFourthOrderStep(xpre, y1pre, y2pre)
        xpre, y1pre, y2pre=xfut, y_1_fut, y_2_fut
    return xfut, y_1_fut, y_2_fut
```

```python
#potential library returns whole arrays instead of single values
def window(xvec,d,cent):
    return 0.5*((np.sign(xvec+d/2.0-cent))-(np.sign(xvec-d/2.0-cent)))

def Vpike(A,xvec,d,cent):
    return A*((1-2*(xvec-cent)/d)*window(xvec,d/2,cent+d/4)+(1+2*(xvec-cent)/d)*window(xvec,d/2,cent-d/4))

def Vsq(A,xvec,d,cent):
    return A*window(xvec,d,cent)

def VJar(A1,A2,A3, xvec,cent):
    return A1*(xvec**4 - 0.5*(A2*(xvec)**2*(np.sign( (xvec-cent) )+1)+A3*(xvec)**2*(np.sign( -(xvec-cent) )+1)))

def VJarDer(A1,A2,A3, xvec,cent):
    return A1*(4*xvec**3 - (A2*(xvec)*(np.sign( (xvec-cent) )+1)+A3*(xvec)*(np.sign( -(xvec-cent) )+1)))



def VpikeDer(A,xvec,d,cent):
    return A*((1-2/d)*window(xvec,d/2,cent+d/4)+(1+2/d)*window(xvec,d/2,cent-d/4))


```

```python
def func_prime_1(x, y_1, y_2):
    return y_2
'''
def func_prime_2(x, y_1, y_2):
    return -(0.97*y_1)**81-(Vpike(B*(x/10),y_1+h/2,d,cent)-Vpike(B*(x/10),y_1-h/2,d,cent))/(2*h)

'''
def func_prime_2(x, y_1, y_2, times, B):
    val=0
    W=1.0
    d=0.0125
    cent=0.0
    A=1
    
    
    if x<times:
        
        #F1=-(Vpike(B*(x/times),y_1+tau/2,d,cent)-Vpike(B*(x/times),y_1-tau/2,d,cent))/(2*tau)
        F1=-(VpikeDer(B*(x/times),y_1,2*d,cent))
        Ldes=W #desired final length after first stage
        L=W   #initial length
        L1=0.0 #amount to be substracted in third stage, for Parrondo this is equal to the final length plus d
        a=(Ldes-L+L1)/2  #new position of the 'center of mass' of the trap after first stage
        b=(Ldes-L1)/2 #half the size of the trap
        #c=L1/2 #correction to the position of the 'center of mass' of the trap after third stage
        F2=-((y_1-a)/b )**481
        val=F2+F1
        
    if (times<=x) & (2*times>=x):
        #F1=-(Vpike(B,y_1+tau/2,d,cent)-Vpike(B,y_1-tau/2,d,cent))/(2*tau)
        F1=-(VpikeDer(B,y_1,2*d,cent))
        Ldes=W +  (W/2+d)*((x-times)/times) #desired final length after first stage
        L=W   #initial length
        L1=0.0 #amount to be substracted in third stage, for Parrondo this is equal to the final length plus d
        a=(Ldes-L+L1)/2  #new position of the 'center of mass' of the trap after first stage
        b=(Ldes-L1)/2 #half the size of the trap
        #c=L1/2 #correction to the position of the 'center of mass' of the trap after third stage
        F2=-((y_1-a)/b )**481
        val=F2+F1
        
        '''
        F1=(Vpike(B,y_1,d,cent)-Vpike(B,y_1+h,d,cent))/h
        F2=(A*(np.sign(-y_1-W)+1)+A*(np.sign(y_1-(W+(W+d)*((x-10)/10)))+1)-(A*(np.sign(-y_1-h-W)+1)+A*(np.sign(y_1+h-(W+(W+d)*((x-10)/10)))+1)))/h
        val=F1+F2
        '''
    if (x>2*times) & (x<3*times):
        #F1=-(Vpike(B,y_1+tau/2,d,cent)-Vpike(B,y_1-tau/2,d,cent))/(2*tau)
        F1=-(VpikeDer(B,y_1,2*d,cent))
        Ldes=(W+W/2+d) #desired final length after first stage
        L=W   #initial length
        L1=(W/2+d)*((x-2*times)/times) #amount to be substracted in third stage, for Parrondo this is equal to the final length plus d
        a=(Ldes-L+L1)/2  #new position of the 'center of mass' of the trap after first stage
        b=(Ldes-L1)/2 #half the size of the trap
        #c=L1/2 #correction to the position of the 'center of mass' of the trap after third stage
        F2=-((y_1-a)/b )**481
        val=F1+F2
        
    if (x>3*times):
        F1=-(VpikeDer(B,y_1,2*d,cent))
        Ldes=(W + W/2+d) #desired final length after first stage
        L=1   #initial length
        L1=(W/2 + d) #amount to be substracted in third stage, for Parrondo this is equal to the final length plus d
        a=(Ldes-L+L1)/2  #new position of the 'center of mass' of the trap after first stage
        b=(Ldes-L1)/2 #half the size of the trap
        #c=L1/2 #correction to the position of the 'center of mass' of the trap after third stage
        F2=-((y_1-a)/b )**481
        val=F2+F1
    return val 


def RungeKuttaFourthOrderStep_2(x_old, y1_old, y2_old, times, B):
    
    k_1_prime1 = func_prime_1(x_old,y1_old, y2_old)
    k_1_prime2 = func_prime_2(x_old,y1_old, y2_old, times, B)
    
    #first step
    x1 = x_old+ (h/2.0)
    y1_1 = y1_old + (h/2.0) * k_1_prime1
    y2_1 = y2_old + (h/2.0) * k_1_prime2
    k_2_prime1 = func_prime_1(x1, y1_1, y2_1)
    k_2_prime2 = func_prime_2(x1, y1_1, y2_1, times, B)
    
    #second step
    x2 = x_old + (h/2.0)
    y1_2 = y1_old + (h/2.0) * k_2_prime1
    y2_2 = y2_old + (h/2.0) * k_2_prime2
    k_3_prime1 = func_prime_1(x2, y1_2, y2_2)
    k_3_prime2 = func_prime_2(x2, y1_2, y2_2, times, B)
    
    
    #third
    x3 = x_old + h
    y1_3 = y1_old + h * k_3_prime1
    y2_3 = y2_old + h * k_3_prime2
    k_4_prime1 = func_prime_1(x3, y1_3, y2_3)
    k_4_prime2 = func_prime_2(x3, y1_3, y2_3, times, B)
    
    #fourth step
    average_k_1 = (1.0/6.0)*(k_1_prime1 + 2.0*k_2_prime1 + 2.0*k_3_prime1 + k_4_prime1)
    average_k_2 = (1.0/6.0)*(k_1_prime2 + 2.0*k_2_prime2 + 2.0*k_3_prime2 + k_4_prime2)
    
    x_new = x_old + h
    y_1_new = y1_old + h * average_k_1
    y_2_new= y2_old + h * average_k_2
    return x_new, y_1_new, y_2_new
```

```python
#we define the arrays holding the two variables
#lo ideal es tener h a 0.0001 para obtener todos los features de las curvas
h=0.0001
min_x = 0.0

times=100
max_x = 4*times
B=0.5
n_points = int((max_x-min_x)/h)
x = zeros(n_points)
y_1 = zeros(n_points)
y_2 = zeros(n_points)

x[0]   = min_x
y_1[0] = -0.35
y_2[0] = 1.0

for i in range(1,n_points):
    x[i],y_1[i],y_2[i] = RungeKuttaFourthOrderStep_2(x[i-1], y_1[i-1], y_2[i-1], times, B)
```

```python
def func_prime_1(x, y_1, y_2):
    return y_2
'''
def func_prime_2(x, y_1, y_2):
    return -(0.97*y_1)**81-(Vpike(B*(x/10),y_1+h/2,d,cent)-Vpike(B*(x/10),y_1-h/2,d,cent))/(2*h)

'''
def func_prime_2(x, y_1, y_2, times, B, k):
    val=0
    W=1.0
    d=0.0125
    cent=0.0
    A=1
    
    
    if x<times:
        
        #F1=-(Vpike(B*(x/times),y_1+tau/2,d,cent)-Vpike(B*(x/times),y_1-tau/2,d,cent))/(2*tau)
        F1=-(VpikeDer(B*(x/times),y_1,d,cent))
        Ldes=W #desired final length after first stage
        L=W   #initial length
        L1=0.0 #amount to be substracted in third stage, for Parrondo this is equal to the final length plus d
        a=(Ldes-L+L1)/2  #new position of the 'center of mass' of the trap after first stage
        b=(Ldes-L1)/2 #half the size of the trap
        #c=L1/2 #correction to the position of the 'center of mass' of the trap after third stage
        F2=-((y_1-a)/b )**k
        val=F2+F1
        
    if (times<=x) & (2*times>=x):
        #F1=-(Vpike(B,y_1+tau/2,d,cent)-Vpike(B,y_1-tau/2,d,cent))/(2*tau)
        F1=-(VpikeDer(B,y_1,d,cent))
        Ldes=W +  (W/2+d)*((x-times)/times) #desired final length after first stage
        L=W   #initial length
        L1=0.0 #amount to be substracted in third stage, for Parrondo this is equal to the final length plus d
        a=(Ldes-L+L1)/2  #new position of the 'center of mass' of the trap after first stage
        b=(Ldes-L1)/2 #half the size of the trap
        #c=L1/2 #correction to the position of the 'center of mass' of the trap after third stage
        F2=-((y_1-a)/b )**k
        val=F2+F1
        
        '''
        F1=(Vpike(B,y_1,d,cent)-Vpike(B,y_1+h,d,cent))/h
        F2=(A*(np.sign(-y_1-W)+1)+A*(np.sign(y_1-(W+(W+d)*((x-10)/10)))+1)-(A*(np.sign(-y_1-h-W)+1)+A*(np.sign(y_1+h-(W+(W+d)*((x-10)/10)))+1)))/h
        val=F1+F2
        '''
    if (x>2*times) & (x<3*times):
        #F1=-(Vpike(B,y_1+tau/2,d,cent)-Vpike(B,y_1-tau/2,d,cent))/(2*tau)
        F1=-(VpikeDer(B,y_1,d,cent))
        Ldes=(W+W/2+d) #desired final length after first stage
        L=W   #initial length
        L1=(W/2+d)*((x-2*times)/times) #amount to be substracted in third stage, for Parrondo this is equal to the final length plus d
        a=(Ldes-L+L1)/2  #new position of the 'center of mass' of the trap after first stage
        b=(Ldes-L1)/2 #half the size of the trap
        #c=L1/2 #correction to the position of the 'center of mass' of the trap after third stage
        F2=-((y_1-a)/b )**k
        val=F1+F2
        
    if (x>3*times):
        F1=-(VpikeDer(B,y_1,d,cent))
        Ldes=(W + W/2+d) #desired final length after first stage
        L=1   #initial length
        L1=(W/2 + d) #amount to be substracted in third stage, for Parrondo this is equal to the final length plus d
        a=(Ldes-L+L1)/2  #new position of the 'center of mass' of the trap after first stage
        b=(Ldes-L1)/2 #half the size of the trap
        #c=L1/2 #correction to the position of the 'center of mass' of the trap after third stage
        F2=-((y_1-a)/b )**k
        val=F2+F1
    return val 


def RungeKuttaFourthOrderStep_5(x_old, y1_old, y2_old, times, B, k):
    
    k_1_prime1 = func_prime_1(x_old,y1_old, y2_old)
    k_1_prime2 = func_prime_2(x_old,y1_old, y2_old, times, B,k)
    
    #first step
    x1 = x_old+ (h/2.0)
    y1_1 = y1_old + (h/2.0) * k_1_prime1
    y2_1 = y2_old + (h/2.0) * k_1_prime2
    k_2_prime1 = func_prime_1(x1, y1_1, y2_1)
    k_2_prime2 = func_prime_2(x1, y1_1, y2_1, times, B,k)
    
    #second step
    x2 = x_old + (h/2.0)
    y1_2 = y1_old + (h/2.0) * k_2_prime1
    y2_2 = y2_old + (h/2.0) * k_2_prime2
    k_3_prime1 = func_prime_1(x2, y1_2, y2_2)
    k_3_prime2 = func_prime_2(x2, y1_2, y2_2, times, B,k)
    
    
    #third
    x3 = x_old + h
    y1_3 = y1_old + h * k_3_prime1
    y2_3 = y2_old + h * k_3_prime2
    k_4_prime1 = func_prime_1(x3, y1_3, y2_3)
    k_4_prime2 = func_prime_2(x3, y1_3, y2_3, times, B,k)
    
    #fourth step
    average_k_1 = (1.0/6.0)*(k_1_prime1 + 2.0*k_2_prime1 + 2.0*k_3_prime1 + k_4_prime1)
    average_k_2 = (1.0/6.0)*(k_1_prime2 + 2.0*k_2_prime2 + 2.0*k_3_prime2 + k_4_prime2)
    
    x_new = x_old + h
    y_1_new = y1_old + h * average_k_1
    y_2_new= y2_old + h * average_k_2
    return x_new, y_1_new, y_2_new
```

```python
#potential library returns whole arrays instead of single values
def window(xvec,d,cent):
    return 0.5*((np.sign(xvec+d/2.0-cent))-(np.sign(xvec-d/2.0-cent)))

def Vpike(A,xvec,d,cent):
    return A*((1-2*(xvec-cent)/d)*window(xvec,d/2,cent+d/4)+(1+2*(xvec-cent)/d)*window(xvec,d/2,cent-d/4))

def Vsq(A,xvec,d,cent):
    return A*window(xvec,d,cent)

def VJar(A1,A2,A3, xvec,cent):
    return A1*(xvec**4 - 0.5*(A2*(xvec)**2*(np.sign( (xvec-cent) )+1)+A3*(xvec)**2*(np.sign( -(xvec-cent) )+1)))

def VJarDer(A1,A2,A3, xvec,cent):
    return A1*(4*xvec**3 - (A2*(xvec)*(np.sign( (xvec-cent) )+1)+A3*(xvec)*(np.sign( -(xvec-cent) )+1)))



def VpikeDer(A,xvec,d,cent):
    return A*((1-2/d)*window(xvec,d/2,cent+d/4)+(1+2/d)*window(xvec,d/2,cent-d/4))



```

```python
def func_prime_13(x, y_1, y_2):
    return y_2
'''
def func_prime_2(x, y_1, y_2):
    return -(0.97*y_1)**81-(Vpike(B*(x/10),y_1+h/2,d,cent)-Vpike(B*(x/10),y_1-h/2,d,cent))/(2*h)

'''
def func_prime_23(x, y_1, y_2, times, B):
    val=0

    cent=0.0
    A1=1
    A2=B
    A3=B
    
    
    if x<times:
        
        val=-VJarDer(A1,A2*((x-0*times)/times), 0, y_1,cent)
        
    if (times<=x) & (2*times>=x):
        
        val=-VJarDer(A1,A2,A3*((x-times)/times), y_1,cent)
        
        
    if (x>2*times) & (x<3*times):
        
        val=-VJarDer(A1,A2*(1 - (x-2*times)/times),A3, y_1,cent)
        
        
    if (x>3*times)&(x<4*times):

        val=-VJarDer(A1,0,A3*( 1 - (x-3*times)/times), y_1,cent)
        
        
    if (x>4*times):
        
        val=-VJarDer(A1,0,0, y_1,cent)
        
        
    return val 


def RungeKuttaFourthOrderStep_3(x_old, y1_old, y2_old, times, B):
    
    k_1_prime1 = func_prime_13(x_old,y1_old, y2_old)
    k_1_prime2 = func_prime_23(x_old,y1_old, y2_old, times, B)
    
    #first step
    x1 = x_old+ (h/2.0)
    y1_1 = y1_old + (h/2.0) * k_1_prime1
    y2_1 = y2_old + (h/2.0) * k_1_prime2
    k_2_prime1 = func_prime_13(x1, y1_1, y2_1)
    k_2_prime2 = func_prime_23(x1, y1_1, y2_1, times, B)
    
    #second step
    x2 = x_old + (h/2.0)
    y1_2 = y1_old + (h/2.0) * k_2_prime1
    y2_2 = y2_old + (h/2.0) * k_2_prime2
    k_3_prime1 = func_prime_13(x2, y1_2, y2_2)
    k_3_prime2 = func_prime_23(x2, y1_2, y2_2, times, B)
    
    
    #third
    x3 = x_old + h
    y1_3 = y1_old + h * k_3_prime1
    y2_3 = y2_old + h * k_3_prime2
    k_4_prime1 = func_prime_13(x3, y1_3, y2_3)
    k_4_prime2 = func_prime_23(x3, y1_3, y2_3, times, B)
    
    #fourth step
    average_k_1 = (1.0/6.0)*(k_1_prime1 + 2.0*k_2_prime1 + 2.0*k_3_prime1 + k_4_prime1)
    average_k_2 = (1.0/6.0)*(k_1_prime2 + 2.0*k_2_prime2 + 2.0*k_3_prime2 + k_4_prime2)
    
    x_new = x_old + h
    y_1_new = y1_old + h * average_k_1
    y_2_new= y2_old + h * average_k_2
    return x_new, y_1_new, y_2_new
```
